```python
import pandas as pd
df = pd.read_csv("Outliers_z-score.csv")
df
```

```python
df.drop(columns= "Unnamed: 0",inplace= True)
```

### Calculate the Z-scores for the 'height' column and identify outliers.

```python
upper_limit = df["height"].mean() + 3 * df["height"].std()
lower_limit = df["height"].mean() - 3 * df["height"].std()
upper_limit , lower_limit
```

```python
df[(df["height"]>=upper_limit)|(df["height"]<=lower_limit)]
                                
```

### Remove the outliers from the 'height' column based on Z-scores and display the cleaned column.

```python
df_height = df[(df["height"]<=upper_limit )&(df["height"]>=lower_limit)]
df_height
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
# before removal outliers

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df["height"])
plt.title("before removal outliers")

# after removal ourliers
plt.subplot(1,2,2)
sns.distplot(df_height["height"])
plt.title("after removal outliers")
            
plt.show()
```

```python
Q1 = df["performance"].quantile(0.25)
Q3 = df["performance"].quantile(0.75)
iqr = Q3 - Q1
iqr
```

```python
upper_limit = Q3 + 1.5 * iqr
lower_limit = Q1 - 1.5 * iqr
upper_limit , lower_limit
```

```python
df[(df["performance"]>= upper_limit)|(df["performance"]<= lower_limit)]
```

### Remove the outliers from the 'performance' column based on IQR and display the cleaned column.

```python
df_per = df[(df["performance"]<= upper_limit)&(df["performance"]>= lower_limit)]
df_per
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(df["performance"])
plt.title("before removal outliers")

# after removal ourliers
plt.subplot(1,2,2)
sns.boxplot(df_per["performance"])
plt.title("after removal outliers")
            
plt.show()
```
